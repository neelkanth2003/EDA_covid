```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns




covid2 = virus.groupby('country').sum()['New cases'].sort_values(ascending=False).head(10)
covid3 = virus.groupby('country').sum()['New deaths'].sort_values(ascending=False).head(10)
covid4 = virus.groupby('country').sum()['New recovered'].sort_values(ascending=False).head(10)


df = pd.DataFrame({
    'Country': covid2.index, 
    'New Cases': covid2.values, 
    'New Deaths': covid3.values, 
    'New Recovered': covid4.values
})


df_melted = df.melt(id_vars='Country', var_name='Category', value_name='Count')


plt.figure(figsize=(14, 7))
sns.barplot(x='Country', y='Count', hue='Category', data=df_melted, palette=['black', 'red', 'green'])


plt.xlabel("Country")
plt.ylabel("Count")
plt.title("Comparison of New Cases, New Deaths, and New Recoveries by Country")
plt.xticks(rotation=45) 
plt.legend(title="Category")


plt.show()
```

```python
covids = virus.groupby('country')['Deaths'].sum().sort_values(ascending = False).head(5)
```

```python
print(covids)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=covids.values, y=covids.index, palette='Reds')
plt.xlabel('Total Deaths')
plt.ylabel('Country')
plt.title('Top 20 Countries by COVID-19 Deaths')
plt.tight_layout()
plt.xticks(rotation=95) 
plt.show()
```

```python

plt.figure(figsize=(10, 4))
plt.hist(covids.values, bins=30, color='blue', edgecolor='black', alpha=0.9)
plt.xlabel('Total Deaths')
plt.ylabel('Frequency')
plt.title('Histogram of COVID-19 Deaths in Top 20 Countries')

# Show plot
plt.show()
```
